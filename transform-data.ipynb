```python
import pandas as pd
india_weather = pd.DataFrame({
    'city': ['Bangalore', 'Hyderabad','Delhi','Chennai'],
    'temperature':[45, 42, 33, 48],
    'humidity':[55, 52, 70, 49]
})
india_weather
```

```python
usa_weather = pd.DataFrame({
    'city': ['Kansas', 'Maryland','New york','Orlando'],
    'temperature':[15, 23, 25, 30],
    'humidity':[80, 70, 70, 55]
})
usa_weather
```

```python
df = pd.concat([india_weather, usa_weather], ignore_index=True)
df
```

```python
df = pd.concat([india_weather, usa_weather], keys=['india', 'us'])
df
```

```python
temperature_df = pd.DataFrame({
    'city':['Kansas', 'Maryland', 'Mexico', 'Orlando', 'ManHattan'],
    'temperature':[22, 25, 55, 39, 42]
}, index=[0,1,2,3])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city':['Mexico', 'Maryland', 'Kansas', 'New Jersey'],
    'windspeed':[6, 8, 12, 5]
}, [2,1,0,3])
windspeed_df
```

```python
new_df = pd.concat([temperature_df, windspeed_df], axis=1)
new_df
```

```python
s = pd.Series(['Humid', 'cool', 'Dry'], name='event')
s
```

```python
df = pd.concat([temperature_df, s], axis=1)
df
```

```python
merged_df = pd.merge(temperature_df, windspeed_df, on='city')
merged_df
```

```python
merged_df = pd.merge(temperature_df, windspeed_df, on='city', how='outer',indicator=True)
merged_df
```

```python
merged_df = pd.merge(temperature_df, windspeed_df, on='city', how='left')
merged_df
```

```python
merged_df = pd.merge(temperature_df, windspeed_df, on='city', how='right')
merged_df
```

```python
newyork_df = pd.DataFrame({
    'date':['05-01-2023', '05-02-2023','05-03-2023', '05-04-2023'],
    'city':['New york', 'New york', 'New york', 'New york'],
    'temperature':[33, 34, 32, 29],
    'humidity':[56, 52, 55, 60]
})

newyork_df
```

```python
berlin_df = pd.DataFrame({
    'date':['05-01-2023', '05-02-2023','05-03-2023', '05-04-2023'],
    'city':['Berlin', 'Berlin', 'Berlin', 'Berlin'],
    'temperature':[30, 27, 31, 29],
    'humidity':[50, 70, 58, 59]
})

berlin_df
```

```python
nairobi_df = pd.DataFrame({
    'date':['05-01-2023', '05-02-2023','05-03-2023', '05-04-2023'],
    'city':['Nairobi', 'Nairobi', 'Nairobi', 'Nairobi'],
    'temperature':[40, 44, 52, 39],
    'humidity':[65, 55, 33, 57]
})

nairobi_df
```

```python
df = pd.concat([newyork_df, berlin_df, nairobi_df])
df
```

```python
transformed_df = df.pivot(index='city', columns='date')
transformed_df
```

```python
table_df = df.pivot_table(index='city', columns='date', aggfunc='sum', margins=True)
table_df
```

```python
weather3_df = pd.DataFrame({
    'date':['05/01/2023', '05/02/2023', '05/03/2023', '05/04/2023', '12/01/2023', '12/02/2023', '12/03/2023', '12/04/2023'],
    'city':['Berlin', 'Berlin', 'Berlin', 'Berlin', 'Berlin', 'Berlin', 'Berlin', 'Berlin'],
    'temperature':[23, 25, 20, 32, 35, 42, 36, 34],
    'humidity':[66, 69, 75, 60, 74, 59, 53, 61]
})
weather3_df
```

```python
weathermelt_df = pd.DataFrame({
    'day':['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'chicago':[23, 25, 32, 40, 24, 27, 31],
    'chennai':[41, 45, 50, 39, 28, 34, 49],
    'berlin':[28, 34, 33, 29, 24, 31, 21]
})
weathermelt_df
```
